```python

import zipfile

local_dataset_zip = '/content/dataset-buah.zip'
zip_buka = zipfile.ZipFile(local_dataset_zip, 'r')
zip_buka.extractall('/content')
zip_buka.close()
```

```python
import tensorflow as tf
import os
import random

# Path ke direktori dataset
dataset_dir = '/content/Dataset_Buah'

# Daftar kelas
classes = os.listdir(dataset_dir)
num_classes = len(classes)

# Membuat list kosong untuk menyimpan path gambar dan label
image_paths = []
labels = []

# Mendapatkan path gambar dan label dari setiap kelas
for class_id, class_name in enumerate(classes):
    class_dir = os.path.join(dataset_dir, class_name)
    image_filenames = os.listdir(class_dir)
    image_paths.extend([os.path.join(class_dir, filename) for filename in image_filenames])
    labels.extend([class_id] * len(image_filenames))

# Menggabungkan image_paths dan labels menjadi satu list
data = list(zip(image_paths, labels))

# Mengacak data secara acak
random.shuffle(data)

# Jumlah data
num_data = len(data)

# Proporsi pembagian data
train_ratio = 0.6
val_ratio = 0.2

# Jumlah data untuk masing-masing subset
num_train = int(num_data * train_ratio)
num_val = int(num_data * val_ratio)
num_test = num_data - num_train - num_val

# Memisahkan data menjadi subset training, validation, dan test
train_data = data[:num_train]
val_data = data[num_train:num_train+num_val]
test_data = data[num_train+num_val:]

# Mendefinisikan fungsi untuk membaca gambar dan label
def load_image(image_path, label):
    image = tf.io.read_file(image_path)
    image = tf.image.decode_jpeg(image, channels=3)
    image = tf.image.resize(image, (224, 224))
    image = image / 255.0
    return image, label

# Membuat dataset dari subset training, validation, dan test

```
