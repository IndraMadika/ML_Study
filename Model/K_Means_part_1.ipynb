Untuk model K-Means Clustering, beberapa hal yang dibutuhkan adalah:

Data: K-Means Clustering membutuhkan data yang akan dikelompokkan menjadi beberapa cluster. Data ini bisa berupa matriks atau array numerik.

Jumlah Cluster (K): Anda perlu menentukan jumlah cluster yang diinginkan sebelum melakukan training model. Jumlah cluster ini akan mempengaruhi hasil kelompokan data.

Inisialisasi Pusat Cluster: Awalnya, model K-Means akan memilih secara acak beberapa titik sebagai pusat cluster awal. Jumlah titik ini sesuai dengan jumlah cluster yang telah ditentukan.

Metode Perhitungan Jarak: Model K-Means menggunakan metode perhitungan jarak (biasanya jarak Euklides) untuk mengukur kedekatan antara data dengan pusat cluster.

Metode Pembaruan Pusat Cluster: Setelah menghitung jarak antara data dengan pusat cluster, model akan memperbarui posisi pusat cluster dengan memperhitungkan rata-rata data dalam setiap cluster.

Kriteria Berhenti: Anda perlu menentukan kriteria berhenti untuk proses iterasi model. Misalnya, ketika pusat cluster tidak berubah lagi atau jumlah iterasi yang telah ditentukan telah tercapai.

Visualisasi Hasil: Setelah model K-Means selesai dilatih, Anda dapat melakukan visualisasi hasil kelompokan data dengan memplot data dan pusat cluster di grafik.

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans

# Generate random data
np.random.seed(0)
X = np.random.randn(100, 2)  # Contoh data 2D

# Inisialisasi dan latih model K-Means dengan 3 cluster
kmeans = KMeans(n_clusters=3)
kmeans.fit(X)

# Prediksi label cluster untuk setiap data
labels = kmeans.predict(X)

# Dapatkan posisi pusat cluster
centers = kmeans.cluster_centers_

# Visualisasi hasil clustering
plt.scatter(X[:, 0], X[:, 1], c=labels)
plt.scatter(centers[:, 0], centers[:, 1], marker='X', color='red')
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')
plt.title('K-Means Clustering')
plt.show()
```
