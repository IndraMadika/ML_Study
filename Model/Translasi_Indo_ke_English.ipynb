```python
import tensorflow as tf
from tensorflow.keras.preprocessing.text import Tokenizer
from tensorflow.keras.preprocessing.sequence import pad_sequences
from sklearn.model_selection import train_test_split

# Data Training (Contoh Data Bahasa Indonesia dan Terjemahan Bahasa Inggris)
indonesian_texts = ['Selamat pagi', 'Terima kasih', 'Saya suka makan', 'Bagaimana kabarmu']
english_texts = ['Good morning', 'Thank you', 'I like to eat', 'How are you']

# Tokenisasi dan Pembuatan Vocab
tokenizer = Tokenizer()
tokenizer.fit_on_texts(indonesian_texts + english_texts)
indonesian_sequences = tokenizer.texts_to_sequences(indonesian_texts)
indonesian_padded = pad_sequences(indonesian_sequences, padding='post')

english_sequences = tokenizer.texts_to_sequences(english_texts)
english_padded = pad_sequences(english_sequences, padding='post')

# Split Data Training dan Data Validasi
X_train, X_val, y_train, y_val = train_test_split(indonesian_padded, english_padded, test_size=0.2, random_state=42)

# Model Penerjemah (Contoh: Recurrent Neural Network)
model = tf.keras.Sequential([
    tf.keras.layers.Embedding(len(tokenizer.word_index) + 1, 64, input_length=X_train.shape[1]),
    tf.keras.layers.GRU(64),
    tf.keras.layers.RepeatVector(y_train.shape[1]),
    tf.keras.layers.GRU(64, return_sequences=True),
    tf.keras.layers.Dense(len(tokenizer.word_index) + 1, activation='softmax')
])

# Compile Model
model.compile(loss='sparse_categorical_crossentropy', optimizer='adam', metrics=['accuracy'])

# Latih Model
model.fit(X_train, y_train, validation_data=(X_val, y_val), epochs=10, batch_size=1)

# Evaluasi Model dengan Data Validasi
loss, accuracy = model.evaluate(X_val, y_val)
print(f'Loss: {loss}, Accuracy: {accuracy}')
```

```python
import tensorflow as tf
from tensorflow.keras.preprocessing.text import Tokenizer
from tensorflow.keras.preprocessing.sequence import pad_sequences

# Data Training (Contoh Data Bahasa Indonesia dan Terjemahan Bahasa Inggris)
indonesian_texts = ['Selamat pagi', 'Terima kasih', 'Saya suka makan', 'Bagaimana kabarmu']
english_texts = ['Good morning', 'Thank you', 'I like to eat', 'How are you']
testing_texts = ['Selamat pagi']

# Tokenisasi dan Pembuatan Vocab
tokenizer = Tokenizer()
tokenizer.fit_on_texts(indonesian_texts + english_texts)
indonesian_sequences = tokenizer.texts_to_sequences(indonesian_texts)
indonesian_padded = pad_sequences(indonesian_sequences, padding='post')

english_sequences = tokenizer.texts_to_sequences(english_texts)
english_padded = pad_sequences(english_sequences, padding='post')

testing_sequences = tokenizer.texts_to_sequences(testing_texts)
testing_padded = pad_sequences(testing_sequences, padding='post', maxlen=indonesian_padded.shape[1])

# Model Penerjemah (Contoh: Recurrent Neural Network)
model = tf.keras.Sequential([
    tf.keras.layers.Embedding(len(tokenizer.word_index) + 1, 64, input_length=indonesian_padded.shape[1]),
    tf.keras.layers.GRU(64),
    tf.keras.layers.RepeatVector(english_padded.shape[1]),
    tf.keras.layers.GRU(64, return_sequences=True),
    tf.keras.layers.TimeDistributed(tf.keras.layers.Dense(len(tokenizer.word_index) + 1, activation='softmax'))
])

# Compile Model
model.compile(loss='sparse_categorical_crossentropy', optimizer='adam', metrics=['accuracy'])

# Latih Model
model.fit(indonesian_padded, english_padded, epochs=100, batch_size=1)

# Prediksi dengan Data Testing
predicted_sequences = model.predict(testing_padded)
predicted_texts = tokenizer.sequences_to_texts(predicted_sequences.argmax(axis=-1))

# Tampilkan Hasil Prediksi
print(f'Testing Text: {testing_texts[0]}')
print(f'Predicted Text: {predicted_texts}')
```
